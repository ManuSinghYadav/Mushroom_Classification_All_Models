```python
x = a.drop(['class'], axis=1)
y = a['class']
```

####Encoding the Categorical columns (**OneHotEncoder**)

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
x_transformed = ohe.fit_transform(x)
```

```python
x_transformed
```

As this is in "sparse matrix", we will transform it into an array and then to a dataframe.

```python
x_transformed_dense_array = x_transformed.toarray()
```

```python
x_transformed = pd.DataFrame(x_transformed_dense_array)
```

```python
x_transformed
```

###Balancing the dataset.
As the dataset is imbalenced. Lets balence it by using **SMOTE** technique.

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

S = SMOTE()
os_x, os_y = S.fit_resample(x_transformed, y)

#Here os represents "over-sampled"
```

```python
os_y.value_counts()
```

```python
os_x.shape
```

###**Feature Selection**
As now we have got 117 features (columns). Let's reduce them by using **RFE** technique.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

rfe = RFE(estimator = LogisticRegression())
rfe = rfe.fit(os_x, os_y)
```

```python
os_x.columns[rfe.support_]
```

```python
len(os_x.columns[rfe.support_])
```

```python
x = os_x[os_x.columns[rfe.support_]]
```

```python
x
```

###**Train-Test Split**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, os_y, test_size=0.25)
```

##Training (ML Models)

###1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR=LogisticRegression().fit(x_train,y_train)
```

###2. K-Nearest Neighbor (K-NN)

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=5).fit(x_train, y_train)
```

###3. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

SVM = SVC(kernel='linear').fit(x_train, y_train)
```

###4. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB().fit(x_train, y_train)
```

###5. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier().fit(x_train, y_train)
```

###6. Random Forest (Bagging)

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0).fit(x_train, y_train)
```

###7. XGBoost (Boosting)

```python
from xgboost import XGBClassifier

XGB = XGBClassifier().fit(x_train, y_train)
```

##Testing / Model Scores

```python
print("Model Scroes,\n1. Logistic Regression: {}\n2. KNN: {}\n3. SVM: {}\n4. Naive Bayes: {}\n5. Decision Tree: {}\n6. Random Forest: {}\n7. XGBoost: {}".format(LR.score(x_test,y_test), KNN.score(x_test,y_test), SVM.score(x_test,y_test), GNB.score(x_test,y_test), DT.score(x_test,y_test), RF.score(x_test,y_test), XGB.score(x_test,y_test)))
```
